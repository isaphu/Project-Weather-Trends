```python
from matplotlib import pyplot as plt
import pandas as pd

BvGTemp = pd.read_csv('BvGTemp.csv')
print(BvGTemp.head())
```

```python
#line 1 plot
x1 = BvGTemp.Year
y1 = BvGTemp[('7 year City MA')]
plt.plot(x1, y1, label = "Bangkok Temp.")
#line 2 plot
x2 = BvGTemp.Year
y2 = BvGTemp[('7 year Global MA')]
plt.plot(x2, y2, label = "Global Temp.")
# Set the x axis label
plt.xlabel('Year')
# Set the y axis label
plt.ylabel('Degree Celcius')
# Set a title of the current axes.
plt.title('Bangkok Vs. Global Temperature')
# show a legend on the plot
plt.legend()
# Display the figure.
plt.show()
```
